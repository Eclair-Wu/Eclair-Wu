```python
from pyscf import gto,scf
import numpy as np
mol = gto.M(
    atom = '''
F 0 0 0
H 0 0 1.1''',
    charge=1,
    spin=1,
    basis = 'ccpvdz',
)
s=mol.intor('int1e_ovlp')
e1=mol.intor('int1e_nuc')+mol.intor('int1e_kin')
e2=mol.intor('int2e')
nuc=mol.energy_nuc()
mf = scf.UHF(mol)
mf.kernel()
print(mf.mo_energy)
import HatreeFock as HF
f=HF.UHF(e1, e2, s, nuc, mol.nelectron, mol.nao)
f.scf(maxcycle=60,delta=1e-10)
print(f.mo_ene)



```
